```python
sns.boxplot(x="size",y="tip",data=tips)
```

The horizontal lines above and below boxes connected by vertical lines are max and min values(excludes outliers)

The bottom, middle and top of a box is 1st(25%) quartile, median(50%),3rd(75%) quartile and most of values lie in-between 1st and 3rd quartiles

The rhombus shapes also called as outliers are values that have too much deviation from most of values/ that are too far(towards zero or towards infinity)


Ex: 1,2,3,80 - here 80 is outlier

```python
tips.corr() 
```

correlation. -1 <= correlation <=1


Threshold generally set for correlation
Assume 0.5


if correlation between any two cols

 (> 0.5) - high positive coorelation (direct positive proportionality)


 (< 0.5) - partially correlated


 (> -0.5) - partially negatively correlated

 
 (< -0.5) - negatively highly correlated (direct inverse proportionality)
