```python
import matplotlib as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.pyplot.plot(x,y,'r') #'r' - color
plt.pyplot.xlabel("x values")
plt.pyplot.ylabel("y values")
plt.pyplot.title("graph")
plt.pyplot.show()
```

```python
plt.pyplot.scatter(x,y,color = "red",marker="*") 
plt.pyplot.xlabel("x values")
plt.pyplot.ylabel("y values")
plt.pyplot.title("graph")
plt.pyplot.show()
```

To plot graphs in same rows or as subplots

```python
plt.pyplot.subplot(1,2,1) #subplot(no of rows, no of columns, each graph number or col number)
plt.pyplot.plot(x,y,'r') #'r' - color
plt.pyplot.xlabel("x values")
plt.pyplot.ylabel("y values")
plt.pyplot.title("graph")
plt.pyplot.subplot(1,2,2)
plt.pyplot.scatter(x,y,color = "green")
plt.pyplot.xlabel("x values")
plt.pyplot.ylabel("y values")
plt.pyplot.title("graph")
plt.pyplot.show()
```

To plot using fig instead of plt

```python
fig = plt.pyplot.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])#[left(x0),bottom(y0),height,width] - in figure not axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b',label="big")
axes.set_xlabel('x-axis')
axes.set_ylabel("y-axis")
axes.set_title("graph")
axes.legend(loc="upper right") #legend's location can also be changed
axes2.plot(x,y,'r',label="small")
axes2.set_xlabel('x-axis')
axes2.set_ylabel("y-axis")
axes2.set_title("graph")
axes2.legend(loc="upper center")
```

```python
import matplotlib.pyplot as plt
plt.bar(x,y,color="green")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
```

```python
labels = ["php","pyp","java","cpp"]
students = [30,45,15,15]
fig,ax1 = plt.subplots(figsize=(5,5)) #figsize(height,width) 
ax1.pie(students,labels=labels,autopct="%0.3f%%",colors=["orange","red","pink","yellow"]) #percentages as pie graph
ax1.legend(labels,loc="upper right")
```

To add different figures AS SUBPLOTS

```python
labels = ["php","pyp","java","cpp"]
students = [30,45,15,15]
fig = plt.figure(figsize=(10,10)) #figsize(height,width) 
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)
ax.pie(students,labels=labels,autopct="%0.3f%%",colors=["orange","red","pink","yellow"]) #percentages as pie graph
ax.legend(labels,loc="upper right")
ax1.bar(x,y)
ax2.scatter(x,y)
ax3.plot(x,y)
```
